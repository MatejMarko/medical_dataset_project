## 1.1. Preprocessing Data for Classification

```python
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.preprocessing import LabelEncoder

def encoded(vect):
    enc = LabelEncoder()
    label_encoder = enc.fit(vect)
    inte_class = label_encoder.transform(label_encoder.classes_)
    t = label_encoder.transform(vect)
    return t

def int_encode_features(data, i):
    
    return encoded(data[:, i])


data_set = df.drop(["Q38", "Q40", "Q48", "Q49", "Q50", "Q54", "Q55"], axis=1)
data = data_set.values

for i in range(data_set.shape[1]):
    data[:, i] = int_encode_features(data, i)
data = data.astype(float)
target = df["Q48"].values
target = encoded(target)
target = target.astype(float)
target_name = "Q48"
x_train, x_test, y_train, y_test = train_test_split(data, 
                                                    target,
                                                    random_state=42,

                                                   )
x_train
```

## Task 2 (Unsupervised Learning) - Charactering people with mental health disorders

## 2.1. Preprocessing Data for Association Rule Mining

**case Yes**

```python
t2_dataset_ini=df

#We are going to reduce the number of data since now we're only going to analyse the answer of who has a mentalhealth disorder
#Q48-yes=1
for i in range(df.shape[0]):
    if df["Q48"][i]!='Yes':
        t2_dataset_ini = t2_dataset_ini.drop([i])
t2_dataset = t2_dataset_ini.drop(["Q38", "Q40", "Q48", "Q49", "Q50", "Q62"], axis=1)      

t2_names = [x for x in t2_dataset.keys()]
#list(names)
#t2_dataset
dummi = t2_dataset_ini["Q49"].str.get_dummies()
dummi_prof = t2_dataset_ini["Q62"].str.get_dummies()
k = []

for i in dummi["Autism (Asperger's)"].index:
    if dummi["Autism (Asperger's)"][i]==1or dummi["Autism"][i] == 1 or dummi["Asperges"][i] == 1 or dummi["Autism spectrum disorder"][i]==1:
        dummi["Autism Spectrum Disorder"][i] = 1
    if dummi["I haven't been formally diagnosed, so I felt uncomfortable answering, but Social Anxiety and Depression."][i] == 1:
        dummi["Anxiety Disorder (Generalized, Social, Phobia, etc)"][i] = 1
        dummi["Mood Disorder (Depression, Bipolar Disorder, etc)"][i] = 1
    if dummi["Depression"][i] == 1:
        dummi["Mood Disorder (Depression, Bipolar Disorder, etc)"][i] = 1
    if dummi["PTSD (undiagnosed)"][i] == 1:
        dummi["Post-traumatic Stress Disorder"][i] == 1
    
dummi = dummi.drop(["Nan", "PTSD (undiagnosed)", "Transgender","Depression", "Autism (Asperger's)","Autism","Asperges","Autism spectrum disorder", "I haven't been formally diagnosed, so I felt uncomfortable answering, but Social Anxiety and Depression."], axis=1)

for i in dummi:
    k.append("Q49_" + str(i))
dummi.columns=k
k=[]
for i in dummi_prof:
    k.append("Q62_" + str(i))
dummi_prof.columns = k
dummie = pd.get_dummies(t2_dataset,columns=t2_names)

dummie = pd.concat([dummie, dummi, dummi_prof], axis=1)
for i in dummie:
    if i.split("_")[1] == "Nan":
        dummie = dummie.drop([i], axis=1)
```

```python
from mlxtend.frequent_patterns import apriori

frequent = apriori(dummie, min_support=0.6, use_colnames=True)
frequent
```

## 2.2. Finding Associations

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent, metric="confidence", min_threshold=0.1)
print(rules.columns)
s = []
for i in range(rules['antecedents'].shape[0]):
    if 'Q49' in str(rules['consequents'][i]):
        s.append((rules['antecedents'][i] , rules['consequents'][i]))
pd.DataFrame(s, columns=('antecedents', 'consequents'))

#association.to_csv('association_rules.csv', sep='\t', encoding='utf-8')
```

## 2.3 Association Rules - Results and Discussion 

We created association rules for everyone that in the moment of the quizz answered they had a mental health disease. We thought that the most interesting rules to be studied would be those whose consequences are mental illnesses and so we restricted the demonstration of the rules created to those cases. 

Initially, we defined several confidence values to verify if with lower values other diseases appeared as consequent, however as this did not happen, we defined the confidence value to 0.7.

We can only find antecedent features for "Mood Disorder" as demonstrated above and these are presented in the table in point "2.2."


## 2.4. Preprocess Data for Clustering

```python
from sklearn.cluster import KMeans

t2_dataset_ini=df
for i in range(df.shape[0]):
    if df["Q48"][i]!='Yes':
        t2_dataset_ini = t2_dataset_ini.drop([i])
t2_dataset = t2_dataset_ini.drop(["Q38", "Q40", "Q48", "Q49", "Q50", "Q62"], axis=1)
```

## 2.5. Finding Groups

```python
kmeans = KMeans(n_clusters=(len(dummi.keys())), random_state=0)
kmeans = kmeans.fit(pd.get_dummies(t2_dataset,columns=t2_names))

clusters = {}
clusters_num_pe = {}
for i in range(len(dummi.keys())):
    clusters[i] = {}
    clusters_num_pe[i] = 0
    
for i in kmeans.labels_:
    clusters_num_pe[i] += 1
d = {}

dummi = dummi.reset_index()
dummi = dummi.drop(["index"], axis=1)
for i in range(dummi.shape[0]):
    for doenca in dummi:
        if dummi[doenca][i] == 1:
            if doenca not in clusters[kmeans.labels_[i]].keys():
                clusters[kmeans.labels_[i]][doenca] = 1 
            else:
                clusters[kmeans.labels_[i]][doenca] += 1
final = {}
pd.set_option('display.notebook_repr_html', True)
for i in clusters.keys():
    final[i] = []
    for doenca in clusters[i]:
        r = clusters[i][doenca] / clusters_num_pe[i]
        #if r >= 0.5:
        final[i].append((doenca, r))
    final[i] = sorted(final[i], key=lambda x: x[1], reverse=True)
    cd=pd.DataFrame.from_dict(final[i])
    cd.columns = [
      "Doença",
      "     % no Cluster"
    ]
    
    print("Cluster " + str(i))
    display(cd)
    print("\n")
```

## 2.6. Clustering - Results and Discussion 

The use of the K-means algorithm resulted in the distinction of 26 clusters (total number of diseases we considered).
Given the results, we studied the probability of a disease being related to the characteristics of each one of the clusters.

Before we defined 26 as the final number of clusters, we tested to increase it to verify if there were increases in the percentages of the diseases in each cluster, however we considered that although for some diseases it was verified, their percentage wasn't sufficiently relevant to draw better conclusions.

In the tables in point "2.5.", Diseases are ordered by their percentage within the cluster. Thus we can observe that the two most common diseases in each of the clusters are always "Mood Disorder" and "Anxiety Disorder".

Thus, and taking into account only this dataset, we can conclude that these two diseases are the most common among people working in Tech, not being able to relate them to a specific type of characteristics. However, the Attention Deficit Hyperactivity Disorder, was highlighted with a percentage of 50% in cluster 18, which may mean that the characteristics of this cluster may be largely related to the presence of this disease

In addition, there were other diseases that stood out, but we did not consider them conclusive since their probability within the clusters were very low.
